```python
# Write a Python program to calculate the area of a trapezoid.

a = int(input("Enter a = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
Area = ((a+b)/2)*h
print(f"Area of trapezoid is {Area}")
```

```python
# Write a Python program to calculate surface volume and area of a cylinder.

r = int(input("Enter r = "))
h = int(input("Enter h = "))
volume = math.pi*r*r*h
area = ((2*math.pi*r*r)+(2*math.pi*r*h))
print(f"Volume = {volume} and area = {area}")
```

```python
#  Write a python program to convert Days into Years, Months and Days. (Ex: if input of Days = 370 then output will be, 
# years=1, months=0 and days = 5)

days = int(input("Enter number of days = "))
years = days//365
months = (days - years *365)//30
day = (days - years * 365 - months * 30)
print(f"years={years}, months={months}, days={day}")
```
