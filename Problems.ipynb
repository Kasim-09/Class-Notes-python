# `Problem 1` :-  WAP to find the happy number, happy number is any positive number when we replaced the number by sum of the square of its digit and repeat the process until the number equals 1 or it looks endlessly is that case it is not a happy number.

```python
for n in range(1,30):
    result = n
    while result!=1 and result!=4:
        digit=0
        sum1=0
        while result>0:
            digit = result%10
            sum1+=digit**2
            result//=10
        result=sum1
    if result == 1:
        print(n,end=" ")
```
