```python
x = -10
if x > 0:
    if x%2 == 0:
        print("its even positive")
    else:
        print("its odd positive")
else:
    print("its non-positive")
```

## Break, Continue and Pass

```python
for i in range(4):
    print(i)
    if i == 3:
        break
else:
    print("loop completed")
```

```python
for i in range(4):
    if i == 3:
        continue
    print(i)
else:
    print("loop completed")
```

```python
#WAP TO FIND SUM OF n NATURAL NUMBER 
num = int(input("Enter number = "))
for i in range(num):
    num = i+num
print(f"The sum is {num}")
```

```python
#WAP TO FIND SUM OF n NATURAL NUMBER by while loop
num = int (input("Enter number = "))
sum=0
while num>0:
    sum = sum+num
    num-=1
print(sum)
```

```python
#WAP PROGRAM TO FIND THE FACTORIAL OF NUMBER 
num = int(input("Enter a number = "))
fact = 1
if num<0:
    print("Negative does not exists")
elif num==0:
    print(1)
else:
    for i in range(1,num+1):
        fact*=i
    print(fact)
```

```python
#FIBONACCI SERIES (0, 1, 1, 2, 3, 5, 8, 13, 21)
nterms = int(input("Enter terms = "))
count=0
n1=0
n2=1
if nterms<=0:
    print("Enter positive terms...")
elif nterms==1:
    print(n1)
else:
    while count<nterms:
        print(n1,end=" ")
        nth = n1+n2
        n1=n2
        n2=nth
        count+=1
    
```
