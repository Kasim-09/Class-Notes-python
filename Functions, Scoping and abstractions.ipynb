# FUNCTION SPECIFICATION
- Typically includes function name, parameters(if any) and return type

```python
def add(x,y):
    '''
    This function add two numbers..
    Parameters:
    x(int): the first number
    y(int): the second number 
    
    Returns 
    int: the sum of two numbers
    
    '''
    
    result = x + y
    return result
```

## POSITIONAL ARGUMENTS 
- These are arguments passed to function in correct positional order 
- The number and position of argumnent must match
- If we change the order the result may/will change
- If we change the order of arguments, we will get error

```python
def sub(x,y):
    return x-y
```

```python
print(sub(4,5))
print(sub(5,4))
print(sub(4,5,6))
```

## KEYWORD ARGUMENTS 
- We can pass argument by keyword
- In above case, the order does not matter
- We can use the mix o0f positional and keyword arguments
- In above case we have one condition, keyboard argument follow positional argument 

```python
def wish(name,msg):
    print("hi",name,msg)
```

```python
wish(name = 'python', msg = 'good morning')
```

```python
# WAP to ask user for 10 inputs and print the max and min values. The user should enter the value until the quit msg is text

# Intialise variables
sum = 0
count = 0
minimum = None 
maximum = None 

while True:
    number = input("Enter number = ")
    if number == 'quit':
        break
    number = int(number)
    
    sum+=number 
    count+=1
    
    if minimum == None or number < minimum:
        minimum = number 
    if maximum == None or number > maximum:
        maximum = number 
average = sum/count
print(sum,average,minimum,maximum)
```

```python
# WAP to reverse a given number 
# input: 3456     output: 6543

num = int(input("Enter a number = "))
revnumber = 0
while num != 0:
    digit = num % 10
    revnumber = revnumber*10 + digit
    num//=10
    print("Reversed number = ", revnumber)
    
    
    
#     python tutor
```
