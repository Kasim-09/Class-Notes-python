# Problem-312

```
 Write a python program to check the validity of a Password.
 Primary conditions for password validation:
 1. Minimum 8 characters.
 2. The alphabet must be between [a-z]
 3. At least one alphabet should be of Upper Case [A-Z]
 4. At least 1 number or digit between [0-9]
 5. At least 1 character from [ _ or @ or $]
 Examples:
 Input: Ram@_f1234
 Output: Valid Password
 Input: Rama_fo$ab
 Output: Invalid Password
 Explanation: Number is missing
 Input: Rama#fo9c
 Output: Invalid Password
 Explanation: Must consist from _ or @ or $
```

````python
def passwordcheck(password):
    u,l,s,d = 0,0,0,0
    if len(password) >=8:
        for i in password:
            if i.isupper():
                u+=1
            if i.islower():
                l+=1
            if i.isdigit():
                d+=1
            if i in "@_$":
                s+=1
        if u>=1 and l>=1 and s>=1 and d>=1 and u+l+s+d == len(password):
            print("valid Password")
        else:
            print("Invalid Password")
    else:
        print("Invalid Password")
````

````python
print(passwordcheck("Ram@1234"))
````

## PROBLEM-313__ SWAP CASE PROBLEM

```
INPUT - mr. eD
OUTPUT - Mr. Ed
```

````python
def casechange(inp):
    temp=""
    for i in inp:
        if i.isupper():
            temp=temp+i.lower()
        elif i.islower():
            temp=temp+i.upper()
        else:
            temp+=i
    return temp
````

````python
casechange("mR. eD")
````

##  WAP to find the sum of positive numbers and -ve numbers using $lambda  f^n$
```

         sum of negative numbers = -6
```

````python
l =  [-2,-4,0,1,2,4]
a = list(filter (lambda x:  x>=0,l))
b = list(filter(lambda x: x<0,l))
print(sum(a))
print(sum(b))

````
