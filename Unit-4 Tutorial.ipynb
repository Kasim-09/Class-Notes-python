## WAP to find the number of spaces, uppercase charcter, lowercase character amd digits in the given user input.

````python
spaceCount = 0
uppercaseCount = 0
lowercaseCount = 0
digitsCount = 0
userInput = input("Enter String - ")
for i in userInput:
    if i.isspace():
        spaceCount=spaceCount+1
    elif i.isupper():
        uppercaseCount=uppercaseCount+1
    elif i.islower():
        lowercaseCount == lowercaseCount+1
    elif i.isdigit():
        digitsCount=digitsCount+1
print(spaceCount)
print(uppercaseCount)
print(lowercaseCount)
print(digitsCount)
        
        
````

## WAP to convert first and last character of a word to uppercase in a give string 

```
input = this is python
output = ThiS IS PythoN

```

````python
s = input("Enter the string ")
s = s.title()
m = s.split()
result = ""
for i in m:
    result+= i[:-1] + [-1].upper() + " "
print(result)
````
